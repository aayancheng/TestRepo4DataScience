# This is a simple Jupyter notebook to illustrate how texts, codes, and graphic outputs can be integrated in single notebook.

### You need to first import a few libraries in addition to base Python to do useful stuff:
- `numpy` is a library to handle matrices and scientific calculation
- `matplotlib` is for plotting graphs
- `pandas` is for handling a data set with rows and columns

```python
import numpy as np
import matplotlib.pyplot as plt
# import pandas as pd

%matplotlib inline
```

### Let's ploting a straight line $y = 2 X + 5$ and a sin curve

```python
# Creating a vector of numbers called x and calculate a linear function on x
x = np.linspace(0, 10, 100)
y = 2 * x  + 5

# plot y against x as a red line
plt.plot(x, y, color = 'red')
```

```python
# Let's try a sin curve
y = 10 * np.sin(x * 2)

# plot y against x as a red line
plt.plot(x, y, color = 'blue')
```

### Now add some random numbers to a linear equation and plot a scatter graph

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x + 5 + rng.randn(50)
plt.scatter(x, y)
```

### With the scattered dots, we will a perform a magic in attempt to recover the straight line $y=2X + 5$ by using linear regression
### For this exericse, we need to import an additional library for machine learning called `scikit-learn`

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 100)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit, color='red')
```

### Finally, we can confirm that the slope and intercept of the fitted line are close to the original one

```python
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
```
